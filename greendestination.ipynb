# Employee Attrition Analysis – Green Destinations

## 1. Objective

Green Destinations has noticed a recent increase in employee attrition. This report aims to:

- Calculate the current attrition rate.
- Explore the relationship between attrition and key factors:
  - Age
  - Years at Company
  - Monthly Income

The insights will help HR identify trends and recommend strategies to reduce attrition.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv("greendestination (1).csv")
df.head()
```

## 2. Attrition Rate

This section calculates the percentage of employees who have left the company.


```python
attrition_rate = df['Attrition'].value_counts(normalize=True).get('Yes', 0) * 100
print(f"Attrition Rate: {attrition_rate:.2f}%")
```

## 3. Age vs Attrition

We analyze whether age plays a role in the likelihood of leaving the company.


```python
plt.figure(figsize=(6, 4))
sns.boxplot(x='Attrition', y='Age', data=df)
plt.title("Age vs Attrition")
plt.show()
```

## 4. Years at Company vs Attrition

This section explores whether employees with fewer years at the company are more likely to leave.


```python
plt.figure(figsize=(6, 4))
sns.boxplot(x='Attrition', y='YearsAtCompany', data=df)
plt.title("Years at Company vs Attrition")
plt.show()
```

## 5. Monthly Income vs Attrition

Here we examine if income level influences attrition.


```python
plt.figure(figsize=(6, 4))
sns.boxplot(x='Attrition', y='MonthlyIncome', data=df)
plt.title("Monthly Income vs Attrition")
plt.show()
```

## 6. Correlation Heatmap

This heatmap shows correlations between age, income, years at the company, and attrition.


```python
df['AttritionFlag'] = df['Attrition'].map({'Yes': 1, 'No': 0})
corr = df[['Age', 'YearsAtCompany', 'MonthlyIncome', 'AttritionFlag']].corr()

plt.figure(figsize=(6, 4))
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title("Correlation with Attrition")
plt.show()
```

## 7. Final Insights & Recommendations

### Insights:
- The attrition rate is approximately **16.12%**.
- Younger employees are more likely to leave the company.
- Employees with fewer years at the company also show higher attrition.
- Lower monthly income may be linked to increased attrition.

### Recommendations:
- Enhance onboarding and support for younger or newer employees.
- Consider retention bonuses or growth opportunities for those early in their career.
- Explore income adjustments where feasible.

This analysis can help HR prioritize areas of concern and take data-driven actions to reduce employee turnover.
